## **Memeriksa Outliers**

Melihat nilai ekstrem (outliers) pada kolom yang bertipe data kan numerik. Hal ini dilakukan agar mendapatkan insight yang akurat sehingga dapat memperkuat analisis dan model yang akan kita bangun nantinya.
<br>

Terdapat Outliers pada kolom Conversion Rate sejumlah 3.4 persen dan kolom lainnya bersih dari outliers.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

# **Eksplorasi Data (EDA)**

## **Comparision (Perbandingan)**

Perbandingan setiap clicks yang di dapatkan dari Device (Perangkat). karena inkonsistensi penulisan pada kolom perangkat membuat beberapa data menjadi terpisah meski data tersebut sama seperti desktop dan Desktop, ini adalah dua perangkat yang sama namun terlihat berbeda dalam grafik karen perbedaan penulisan dan dianggap sebagai variabel yang berbeda.

```python
variabel = df.groupby('Device')['Clicks'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=variabel.index, y=variabel.values, palette='viridis', hue=variabel.index, legend=False)
plt.title('Total Clicks by Device')
plt.xlabel('Device')
plt.ylabel('Total Clicks')
plt.xticks(rotation=45)
plt.show()
```

## **Composition (Komposisi)**

Komposisi dari setiap penjualan yang dihasilkan oleh perangkat, karena penulisan pada perangkat tidak konsisten membuatnya menjadi terpisah beberapa bagian meski termasuk data yang sama.

```python
variabel = df.groupby('Device')['Sale_Amount'].sum().sort_values(ascending=False)
variabel2 = variabel.head(9)

plt.figure(figsize=(10, 8))
variabel2.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Device Composition by Sale Amount')
plt.ylabel('')
plt.axis('equal')
plt.show()
```
