# **Business Understanding**

## **Business Objective**

**1. Tujuan Bisnis**<br>
Tujuan bisnis dari penggunaan data ini adalah untuk memaksimalkan Return on Ad Spend dan efisiensi konversi pemasaran. Perusahaan ingin memastikan bahwa setiap uang yang dikeluarkan untuk iklan Google Ads menghasilkan jumlah Leads dan Sale Amount yang optimal, serta mengidentifikasi perangkat mana yang paling menguntungkan.
<br>
<br>
**2. Deskripsi Dataset**<br>
Dataset ini berisi catatan performa iklan harian berdasarkan berbagai kampanye dan kata kunci. Komponen utamanya meliputi:

- Identitas Iklan: Ad_ID, Campaign_Name, dan Keyword yang digunakan.

- Metrik Performa: Impressions (Berapa kali iklan muncul).

- Clicks (Berapa kali iklan diklik).

- Leads (Calon pelanggan yang masuk).

- Conversions (Transaksi yang berhasil).

- Metrik Keuangan: Cost dan Sale_Amount.

- Dimensi Pendukung: Ad_Date, Location, dan Device.
<br>

**3. Tujuan Analisis Bisnis**<br>
Mengapa kita perlu menganalisis data ini?
- Analysis data biaya: Mengetahui kampanye mana yang memiliki biaya tinggi tetapi konversinya rendah.

- Optimalisasi Transaksi yang Berhasil: Menganalisis korelasi antara Keyword dengan Conversion Rate untuk memperbaiki strategi penargetan.

- Peforma Jaringan dan Perangkat: Menentukan apakah pengguna Mobile lebih cenderung melakukan pembelian dibandingkan pengguna Desktop.

- Trend Analysis: Melihat apakah ada hari-hari tertentu di mana performa iklan melonjak atau menurun drastis.
<br>

**4. Masalah pada Dataset**<br>
Beberapa masalah ada pada Dataset ini yaitu:

- Inkonsistensi Penulisan: Campaign_Name memiliki banyak variasi salah ketik: "Data Anlytics Corse", "Data Analytcis Course", "DataAnalyticsCourse". Ini akan merusak agregasi data.

- Format Tanggal Tidak Seragam: Kolom Ad_Date mencampur format YYYY-MM-DD, DD-MM-YYYY, dan YYYY/MM/DD. Ini akan menyebabkan error saat proses analisis waktu.

- Missing Values: Ada kolom seperti Conversion Rate yang terlihat kosong di beberapa baris awal.

- Tipe Data Berantakan: Kolom Cost mengandung simbol mata uang ($), yang membuatnya terbaca sebagai teks (string), bukan angka. Anda tidak bisa menjumlahkan biaya sebelum simbol ini dibersihkan.
